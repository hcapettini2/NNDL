## The Neural Network

About the CNN dimensions  https://cs231n.github.io/convolutional-networks/#conv

Convolutiona layer https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html

Pooling layer https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html

```python
class ConvNet(nn.Module):
    def __init__(self,activation = nn.ReLU(),n_filters = 16,n_channels = 1,dropout=0.):
        """
        In this case I will use a super simple CNN to understand how to set it properly.
        The architecture is the following:
        
        INPUT -> [[CONV1->RELU]*1->POOL]*2 -> [FC ->RELU] *1 -> FC              
        **FC = Fully Conected
        
        n_channels : # of channels supporting the images (1 for greyscale)
        n_filters_1: # of filters in the first convolution
        n_filters_2: # of filters in the second convolution
        """
        super().__init__()
        
        #Define the convolution - relu - pooling set
        self.cnn_layers = nn.Sequential(
            nn.Conv2d(in_channels  = n_channels,                 #CON
                      out_channels = n_filters,
                      kernel_size  = 5,
                      stride  = 1,
                      padding = 2),
            activation,                                          #ACT
            nn.MaxPool2d(kernel_size = 2,stride = 2),            #POOL
            
            nn.Conv2d(in_channels  = n_filters,                #CON2
                      out_channels = n_filters * 2,
                      kernel_size  = 5,
                      stride  = 1,
                      padding = 2),
            activation,                                          #ACT
            nn.MaxPool2d(kernel_size = 2,stride = 2),            #POOL
        )
        
        #Define a dropout
        self.dropout = nn.Dropout(dropout)
        
        #Define the linear part
        self.linear_layers = nn.Sequential(
            nn.Linear(in_features  = (n_filters * 2) * 7 * 7,     #depth (filters in the last conv) * hight * width 
                      out_features = (n_filters * 2) * 7 * 7),
            activation
        )
        
        #Define the FC output
        self.out = nn.Sequential(
            nn.Linear(in_features = (n_filters * 2) * 7 * 7,
                     out_features = 10)         #Output for 10 categories
        )        
    
    def forward(self,x):
        x = self.cnn_layers(x)
        x = x.view(x.size(0),-1)
        x = self.dropout(x)
        x = self.linear_layers(x)
        x = self.out(x)
        return(x)
    
    
    def losses_plot(train,val, save=False):
        # Plot losses
        plt.figure(figsize=(12,8))
        plt.semilogy(train, label="Training los")
        plt.semilogy(val, label="Validation loss")
        plt.title('Loss', fontsize = 20)
        plt.xlabel('Epoch', fontsize = 15)
        plt.ylabel('Loss', fontsize = 15)
        plt.legend()
        if save: plt.savefig('./imgs/classification/loss.svg', format='svg')
        plt.show()
    
    def model_accuracy(model,test):
       # Input and true label
        x_test,y_test = next(iter(test))

        #Network evaluation
        model.eval()
        with torch.no_grad():
            y_pred = model(x_test).argmax(dim=1)
        
        #Estimate accuracy on test dataset
        a = np.array(y_test)
        b = np.array(y_pred)
        accuracy = np.sum(a==b)/len(a) * 100
        print(f"The obtained accuracy is {accuracy} %")
    
        return y_test,y_pred
    
    def confussion_plot(real, prediction):
        categories = ['T-shirt/top','Trouser','Pullover',
                       'Dress','Coat','Sandal','Shirt',
                       'Sneaker','Bag','Ankle boot']

        array = confusion_matrix(real, prediction,normalize="true")
        df_cm = pd.DataFrame(array, index = categories, columns = categories)
    
        fig = plt.figure(figsize=(8,7))
        sns.heatmap(df_cm, annot=True, fmt='.2f')
        plt.ylabel("Correct")
        plt.xlabel("Predicted")
        plt.title("Confussion matrix")
        plt.show()
        
        
    def weight_plot(trained_model):
        # First conv layer
        c1_w = trained_model.cnn_layers[0].weight.data.cpu().numpy()
        c1_b = trained_model.cnn_layers[0].bias.data.cpu().numpy()

        # Second conv layer
        c2_w = trained_model.cnn_layers[3].weight.data.cpu().numpy()
        c2_b = trained_model.cnn_layers[3].bias.data.cpu().numpy()

        # Linear layer
        h_w = trained_model.linear_layers[0].weight.data.cpu().numpy()
        h_b = trained_model.linear_layers[0].bias.data.cpu().numpy()

        # Output linear layer
        o_w = trained_model.out[0].weight.data.cpu().numpy()
        o_b = trained_model.out[0].bias.data.cpu().numpy()   


        # Weights histogram
        fig, axs = plt.subplots(4, 1, figsize=(14,8))
        axs[0].hist(c1_w.flatten(), 50)
        axs[0].set_title('First convolutional layer weights')
        axs[1].hist(c2_w.flatten(), 50)
        axs[1].set_title('Second convolutional layer weights')
        axs[2].hist(h_w.flatten(), 50)
        axs[2].set_title('Linear layer weights')
        axs[3].hist(o_w.flatten(), 50)
        axs[3].set_title('Output layer weights')
        [ax.grid() for ax in axs]
        plt.tight_layout()
        plt.show()        
        
        
    def plot_filters(trained_model):
        filter1 = trained_model.cnn_layers[0].weight.data.cpu().numpy()
        filter2 = trained_model.cnn_layers[3].weight.data.cpu().numpy()

        #First convolutional layer
        fig, axs = plt.subplots(2,8, figsize=(15, 6))
        axs = axs.ravel()
        for i in range(16):
            axs[i].imshow( filter1[i, 0, :, :], cmap='Greys')
            axs[i].axis('off')

        fig.suptitle('Filters of first convolutional layer')
        plt.show()

        #Second convolutional layer
        fig, axs = plt.subplots(4,8, figsize=(15, 12))
        axs = axs.ravel()
        for i in range(32):
            axs[i].imshow( filter2[i, 0, :, :], cmap='Greys')
            axs[i].axis('off')

        fig.suptitle('Filters of second convolutional layer')
        plt.show()
        
        
```

```python
#def train(net,train_data,val_data,num_epochs=5,l_rate=1e-3,reg_term=0., prints=False):
def train(net,train_data,val_data,optimizer,num_epochs=5, prints=False):    
    """
    
    """
    
    #Define the loss function
    loss_fn = nn.CrossEntropyLoss()
    
    #Define the optimizer
    #optimizer = optim.Adam(net.parameters(), lr =l_rate, weight_decay=reg_term)
    

    train_loss_log = []
    val_loss_log = []
    for epoch_num in range(num_epochs):
        if prints: print(f"EPOCH {epoch_num}")
            
        ###TRAIN
        train_loss = []
        net.to(device)
        net.train()
        
        for sample_batched in train_data:
            #Move data to device
            x_batch = sample_batched[0].to(device)
            label_batch = sample_batched[1].to(device)
            
            #Forward pass
            out = net(x_batch)
            
            #Compute loss
            loss = loss_fn(out, label_batch)
            
            #Back propagation
            net.zero_grad()
            loss.backward()
            
            #Update weights
            optimizer.step()
            
            #Save train loss for this batch
            loss_batch = loss.detach().cpu().numpy()
            train_loss.append(loss_batch)
            
        #Save average train loss
        train_loss = np.mean(train_loss)
        if prints: print(f"AVERAGE TRAIN LOSS: {train_loss}")
        train_loss_log.append(train_loss)
        
        ###VALIDATION
        val_loss = []
        net.eval()
        with torch.no_grad():
            for sample_batched in val_data:
                #Move data to device
                x_batch = sample_batched[0].to(device)
                label_batch = sample_batched[1].to(device)
            
                # Forward pass
                out = net(x_batch)

                # Compute loss
                loss = loss_fn(out, label_batch)
                                   
                # Save val loss for this batch
                loss_batch = loss.detach().cpu().numpy()
                val_loss.append(loss_batch)

            # Save average validation loss
            val_loss = np.mean(val_loss)
            if prints: print(f"AVERAGE VAL LOSS: {np.mean(val_loss)}")
            val_loss_log.append(val_loss)
            
    return train_loss_log,val_loss_log
```

### Grid search

```python
batch_size = 500
train_dataloader      = DataLoader(train_dataset,      batch_size=batch_size,              shuffle=True, num_workers=0)
validation_dataloader = DataLoader(val_dataset, batch_size=len(val_dataset), shuffle=False, num_workers=0)
test_dataloader       = DataLoader(test_dataset,       batch_size=len(test_dataset),       shuffle=False, num_workers=0)
```

```python
#Define the hyperparameters to feed the gridsearch
gs_filters      = [16,32]#[25,50,75,100]
gs_activations  = [nn.ReLU(),nn.Sigmoid()]#[25,50,75,100]
gs_learning_rate= [1e-4,1e-3]
gs_weight_decay = [0,1e-4]
gs_optimization = [optim.Adam,optim.SGD]
gs_dropout      = [0.,0.1]
epochs = 5
max_batch_size = 100

best_loss = 100

#Keep a register for the lowest loss in each model
#We consider the best model the one with the lowes validation loss
m_loss = []
trial = 0
for filters in gs_filters:
    for dr in gs_dropout:
        for act in gs_activations:
            for lr in gs_learning_rate:
                for wd in gs_weight_decay:
                    for c_opt in gs_optimization:
                        net = ConvNet(activation = act,n_filters = filters,dropout=dr)                #Define network
                        opt = c_opt(net.parameters(), lr =lr, weight_decay=wd)               #Define optimizer
                        t_loss,v_loss = train(net,train_dataloader,validation_dataloader,  #Define train the network
                                              num_epochs=epochs,optimizer=opt,prints=True)

                        m_loss.append((trial,v_loss[-1]))
                        trial = trial + 1
                        print("Filters:",filters,"Activation:",act,"lr:",lr,"wd:",wd,"Optimizer:",c_opt,
                              "dropout",dr,"Loss",v_loss[-1])

                        if v_loss[-1] < best_loss:
                            best_loss = v_loss[-1]
                            best_config = {
                                "Filters":filters,
                                "Activation":act,
                                "lr":lr,
                                "wd":wd,
                                "dropout":dr,
                                "Optimizer":c_opt
                            }
                            print("Best model till now:")
                            print(best_config)
```
